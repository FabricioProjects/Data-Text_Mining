**Tokenization**

```python
# pre-processing chain that will consider 'tweets language'
# regular expressions library
emoticons_str = r"""
    (?:
        [:=;] # Eyes
        [oO\-]? # Nose (optional)
        [D\)\]\(\]/\\OpP] # Mouth
    )"""
 
regex_str = [
    emoticons_str,
    r'<[^>]+>', # HTML tags
    r'(?:@[\w_]+)', # @-mentions
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)", # hash-tags
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', # URLs
 
    r'(?:(?:\d+,?)+(?:\.?\d+)?)', # numbers
    r"(?:[a-z][a-z'\-_]+[a-z])", # words with - and '
    r'(?:[\w_]+)', # other words
    r'(?:\S)' # anything else
]
    
tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)
 
def tokenize(s):
    return tokens_re.findall(s)
 
def preprocess(s, lowercase = True):
    tokens = tokenize(s)
    if lowercase:
        tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
    return tokens
```

**Filtering the data**

```python
# tokenizing line by line
tokens = []
with open('/home/teresas/csv_files/token/atletiba.json', 'r') as f:
    for line in f:
        corpus = json.loads(line)
        # Create a list with all the terms
        punctuation = list(string.punctuation)
        stop = stopwords.words('portuguese') + punctuation + \
                ['rt','‚Ä¶','coxa','coritiba','furacao','alviverde','atletiba','atleticopr','atletico','vai','pra',\
                 'atl','√©tico','nao','aqui','√£o','1','q','x','parab','√©ns']
        tokens = tokens + [term for term in preprocess(corpus['comments']) if term not in stop]
```
