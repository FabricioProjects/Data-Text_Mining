```python
# Import the necessary methods from tweepy library
import tweepy
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
import simplejson as json
import pandas as pd
pd.set_option('max_colwidth',500)
```

```python
# accessing the twitter api
consumer_key = '...'
consumer_secret = '...'
access_token = '...'
access_secret = '...'
 
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
# The api variable is our entry point for most of the operations we can perform with Twitter.
api = tweepy.API(auth)
```

```python
# reading our own timeline
for status in tweepy.Cursor(api.home_timeline).items(10):
    # Process a single status
    print(status.user.screen_name)
    print(status.created_at)
    print(status.text) 
```

**Getting filtered search tweets**

```python
# get the stream listener, manipulate the data allocation and filter the search
class MyListener(StreamListener):
 
    def on_data(self, data):
        try:
            with open('/home/teresas/csv_files/twitter/atletiba.txt', 'a') as f:
                f.write(data)
                all_data = json.loads(data)
                print(all_data["text"])
#                print(all_data["user"])
#                print(all_data["screen_name"])
                return True
        except BaseException as e:
            print(str(e))
        return True
 
    def on_error(self, status):
        print(status)
        return True
 
twitter_stream = Stream(auth, MyListener())
twitter_stream.filter(track=['Coxa','coritiba','Coritiba','Alviverde', \
                             'atletiba','Atletiba','ATLETIBA','AtleTiba','CAPxCFC','CFCxCAP', \
                             'Atl√©tico-PR','atleticopr','Furac√£o','furacao'])
```

```python
# read the data in into an array that we call tweets.
tweets_data_path = '/home/teresas/csv_files/twitter/atletiba.txt'
tweets_data = []
tweets_file = open(tweets_data_path, "r")
for line in tweets_file:
    try:
        tweet = json.loads(line)
        tweets_data.append(tweet)
    except:
        continue
        
# printing the number of tweets
print(len(tweets_data))
```

```python
# creating a dataframe
tweets = pd.DataFrame()
```

```python
# some options to fill the dataframe
tweets_data[1:2]
```

```python
# fill up the dataframe with some variables 
tweets['name'] = list(map(lambda tweet: tweet['user']['name'], tweets_data))
tweets['url'] = list(map(lambda tweet: tweet['user']['url'], tweets_data))
tweets['lang'] = list(map(lambda tweet: tweet['lang'], tweets_data))
tweets['description'] = list(map(lambda tweet: tweet['user']['description'], tweets_data))
tweets['created_at'] = list(map(lambda tweet: tweet['created_at'], tweets_data))
tweets['source'] = list(map(lambda tweet: tweet['source'], tweets_data))
tweets['text'] = list(map(lambda tweet: tweet['text'], tweets_data))
tweets['location'] = list(map(lambda tweet: tweet['user']['location'], tweets_data))
tweets['timestamp_ms'] = list(map(lambda tweet: tweet['timestamp_ms'], tweets_data))
tweets['coordinates'] = list(map(lambda tweet: tweet['coordinates'], tweets_data))
```
